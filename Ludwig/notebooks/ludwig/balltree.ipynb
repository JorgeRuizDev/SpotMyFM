```python
import pandas as pd
from sklearn.neighbors import BallTree
from random import randint
```

```python
df = pd.read_csv("../../dataset-tools/python/out/extracted_features.csv", index_col=0).dropna()
```

```python
df.head()
```

```python
print(df.shape)
```

```python
numpy_df = df.drop("0", axis=1).to_numpy()

# scale the data with minmax scaler from sklearn
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler = scaler.fit(numpy_df)
scaled_data = scaler.fit_transform(numpy_df)

print(scaled_data.shape)
# shuffle scaled_data
shuffled_data = scaled_data[randint(0, scaled_data.shape[0]) :, :]
tree = BallTree(shuffled_data)

```

```python
rand_int = randint(0, df.shape[0])
dist , res = tree.query([scaled_data[rand_int ]], k=2)


for i in res:
    print(df.iloc[i]["0"])
```

# Distancia media:
La distancia media entre las filas es de 2606 elementos.
Los datos están ordenados por género, por lo que lo que cuanto menor sea la diferencia, mayor es la "semejanza" respecto al género. 

```python
diff = 0 

for i in range(len(scaled_data)):
    _, res = tree.query([scaled_data[i]], k=2)
    res = res[0]
    diff += abs(res[0] - res[1])
    # print i / len percentage with two decimals
    print(diff / i)
```
